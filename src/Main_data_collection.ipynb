```python
seasons = ['2010-11', '2011-12', '2012-13', '2013-14', '2014-15', '2015-16', 
          '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22', 
          '2022-23', '2023-24', '2024-25']
```

```python
for season in seasons:
    try:
        print(f"Collecting data for {season}...")  
        advanced_stats = leaguedashteamstats.LeagueDashTeamStats(
            season=season,
            measure_type_detailed_defense='Advanced'
        ).get_data_frames()[0]

        advanced_stats.to_csv(f"../data/API_Data/team_stat_{season}.csv", index=True)
        print(f"Successfully saved {season}")  
        time.sleep(2)
        
    except Exception as e:
        print(f"Error for {season}: {e}")
        continue
```
