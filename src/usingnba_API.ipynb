```python
nba_players = players.get_players()
all_players = pd.DataFrame(nba_players)
```

Getting players data

```python
all_players.head()
```

```python
all_players.to_csv("../data/API_Data/all_players.csv", index = True)
```

```python
# to make a dataset only for the active players
active_players = []
for players in nba_players:
    if players["is_active"] == True:
        active_players.append(players) # this will give a list
```

```python
active_players_df = pd.DataFrame(active_players)
active_players_df.to_csv("../data/API_DATA/active_players.csv", index = True)
```
